Sample Test Code

```python
from qiskit import QuantumCircuit
from qiskit.circuit import Parameter
from qiskit.primitives import StatevectorSampler

x = Parameter('x')
theta = Parameter('θ')

qc = QuantumCircuit(1)

# feature map
qc.rx(x, 0)

# ansatz
qc.ry(theta, 0)

qc.measure_all()

# bind values (IMPORTANT)
qc = qc.assign_parameters({x: 0.5, theta: 1.2})

sampler = StatevectorSampler()
result = sampler.run([qc]).result()

counts = result[0].data.meas.get_counts()
print(counts)
```

Deutsch-Jozsa Algorithm

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler

n = 3
qc = QuantumCircuit(n+1)

qc.x(n)
qc.h(range(n+1))

# balanced oracle
for i in range(n):
    qc.cx(i, n)

qc.h(range(n))
qc.measure_all()

sampler = StatevectorSampler()
print(sampler.run([qc]).result()[0].data.meas.get_counts())

print(qc)
```

Bernstein-Vazirani

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler

s = "101"
n = len(s)

qc = QuantumCircuit(n+1)
qc.x(n)
qc.h(range(n+1))

for i, bit in enumerate(s):
    if bit == "1":
        qc.cx(i, n)

qc.h(range(n))
qc.measure_all()

sampler = StatevectorSampler()
print(sampler.run([qc]).result()[0].data.meas.get_counts())

print(qc)
```

Simon's Algorithm

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler

n = 2
qc = QuantumCircuit(2*n)

qc.h(range(n))
qc.cx(0,2)
qc.cx(1,3)

qc.h(range(n))
qc.measure_all()

sampler = StatevectorSampler()
print(sampler.run([qc]).result()[0].data.meas.get_counts())

print(qc)
```

Grover's Algorithm

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler

qc = QuantumCircuit(2)

qc.h([0,1])
qc.cz(0,1)  # mark |11>

qc.h([0,1])
qc.x([0,1])
qc.cz(0,1)
qc.x([0,1])
qc.h([0,1])

qc.measure_all()

sampler = StatevectorSampler()
print(sampler.run([qc]).result()[0].data.meas.get_counts())

print(qc)
```

Quantum Fourier Transform (QFT)

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler
import numpy as np

qc = QuantumCircuit(3)

for j in range(3):
    qc.h(j)
    for k in range(j+1,3):
        qc.cp(np.pi/2**(k-j), k, j)

qc.measure_all()

sampler = StatevectorSampler()
result = sampler.run([qc]).result()
counts = result[0].data.meas.get_counts()
print(counts)

print(qc)
```

Quantum Phase Estimation

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler
import numpy as np

qc = QuantumCircuit(3)

qc.x(2)
qc.h([0,1])

qc.cp(np.pi/2,0,2)
qc.cp(np.pi,1,2)

qc.h(1)
qc.cp(-np.pi/2,0,1)
qc.h(0)

qc.measure_all()

sampler = StatevectorSampler()
print(sampler.run([qc]).result()[0].data.meas.get_counts())

print(qc)
```

VQE (Variational Quantum Eigensolver)

```python
from qiskit.circuit import QuantumCircuit, Parameter
from qiskit.primitives import StatevectorEstimator
import numpy as np

theta = Parameter('θ')

qc = QuantumCircuit(1)
qc.ry(theta, 0)

est = StatevectorEstimator()
job = est.run([(qc, "Z", [np.pi/4])])
print(job.result())

print(qc)
```

QAOA

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler
import numpy as np

qc = QuantumCircuit(2)

qc.h([0,1])
qc.cz(0,1)
qc.rx(np.pi/4, [0,1])

qc.measure_all()

sampler = StatevectorSampler()
result = sampler.run([qc]).result()
counts = result[0].data.meas.get_counts()
print(counts)

print(qc)
```

VQE (Variational Quantum Classifier)

```python
from qiskit.circuit import QuantumCircuit, Parameter
from qiskit.primitives import StatevectorSampler

x = Parameter('x')
theta = Parameter('θ')

qc = QuantumCircuit(1)

qc.rx(x,0)
qc.ry(theta,0)

qc.measure_all()

sampler = StatevectorSampler()

result = sampler.run([(qc, [0.5,1.2])]).result()
counts = result[0].data.meas.get_counts()
print(counts)

print(qc)
```
